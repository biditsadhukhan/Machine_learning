```python
def normal_dist(i, j, river_mean, non_river_mean, river_cov, non_river_cov):
    l = len(img_arr_1[i, j, :])
    r = img_arr_1[i, j, :] - river_mean
    nr = img_arr_1[i, j, :] - non_river_mean
    try:
        river_chol = np.linalg.cholesky(river_cov)
        river_class_d = np.exp(-0.5 * np.sum(np.square(np.linalg.solve(river_chol, r))))
        # density function of river class
        river_class_density = (1 / (((2 * np.pi) ** (l / 2)) * np.prod(np.diagonal(river_chol)))) * river_class_d
        non_river_chol = np.linalg.cholesky(non_river_cov)
        non_river_class_d = np.exp(-0.5 * np.sum(np.square(np.linalg.solve(non_river_chol, nr))))
        # density function of non-river class
        non_river_class_density = (1 / (((2 * np.pi) ** (l / 2)) * np.prod(np.diagonal(non_river_chol)))) * non_river_class_d
    except np.linalg.LinAlgError:
        return 0, 0
    return river_class_density, non_river_class_density
```

```python
def bayes_matrix(prior_r, prior_nr, river_mean, non_river_mean, river_cov, non_river_cov):
    prediction = np.zeros(shape=(512, 512), dtype=np.int64)
    for i in range(512):
        for j in range(512):
            river_density, non_river_density = normal_dist(i, j, river_mean, non_river_mean, river_cov, non_river_cov)
            if river_density * prior_r >= non_river_density * prior_nr:
                prediction[i, j] = 0 # black color represent river classs
            else:
                prediction[i, j] = 1 # white color for nonriver class
    return prediction
```

```python
def bayes_classifier(prior_r, prior_nr,name_river_class, name_non_river_class, river_mean, non_river_mean, river_cov, non_river_cov):
    prediction = np.zeros(shape=(512, 512), dtype=np.int64)
    pred = []
    for i in range(512):
        for j in range(512):
            river_density, non_river_density = normal_dist(i, j, river_mean, non_river_mean, river_cov, non_river_cov)
            if river_density * prior_r >= non_river_density * prior_nr:
                pred.append(1) # white color represent river class
            else:
                  pred.append(0)  # black color for non-river class
#     plt.subplot(1, 1, 1)
#     plt.imshow(prediction, cmap='gray')
#     plt.tick_params(labelleft=False, labelbottom=False, bottom=False, right=False, left=False, which='both')
#     plt.show()
    k = np.array(pred).reshape(512,512)  
    plt.subplot(1,1,1)
    print(f"zero(black color) represent {name_non_river_class} & one(white) represent {name_river_class}")
    plt.imshow(k,cmap = 'gray')
    plt.tick_params(labelleft = False , labelbottom = False , bottom = False ,right = False ,left = False,which = 'both')
   # plt.subplots_adjust(right = 0.8, top = 1.5)
    plt.title(f"{name_river_class} vs {name_non_river_class}")
    plt.show()
```
